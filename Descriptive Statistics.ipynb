# Importance of Statistics

Statistics is a fundamental tool in data science, and it plays a crucial role in various aspects of the data science lifecycle. Here are some of the reasons why statistics is important in data science:

1. **Data Collection**: Statistics provides methods for sampling, experimental design, and survey methodology, which are essential for collecting high-quality data. These methods help ensure that the data collected is representative of the population of interest, and it reduces the risk of bias.

2. **Data Cleaning and Preprocessing:** In real-world data, there can be missing values, outliers, and other anomalies that need to be dealt with before analysis. Statistics provides techniques to identify and handle such issues to ensure that the data is of high quality and appropriate for analysis.

3. **Data Analysis:** Statistics provides a wide range of techniques for analyzing data, including hypothesis testing, regression analysis, clustering, and classification. These techniques are used to extract insights from the data and make informed decisions.

4. **Model Selection and Validation**: Statistics provides methods for selecting appropriate models and validating their performance. This includes techniques such as cross-validation, which is used to estimate the performance of a model on new data.

5. **Visualization**: Statistics provides techniques for visualizing data, which is an essential aspect of data science. Visualization allows us to explore the data, identify patterns, and communicate insights effectively.

In summary, statistics is a critical component of data science that underpins various aspects of the data science lifecycle. It helps to ensure that data is of high quality, provides methods for analyzing data, and enables informed decision-making based on data insights.

#Descriptive statistics

Descriptive Statistics, provides a summary of your dataset giving a measure of the centre, dispersion and shape of your data. Here the data is described as a sample of the whole population, and there are no inferences made from the sample to the whole population, unlike Inferential Statistics, in which we model the data on the basis of probability theory.

## Key Elements of Descriptive Statistics

### Measures Of Central Tendency

* Mean
* Median
* Mode

### Measures Of Spread

* Range
* Outliers
* Interquantile Range
* Variance

### Dependence

* Correlation v/s Causation
* Covariance

```python
#8) Calculate the Interquartile range for the number of school absences.
```

```python
Q1 = df['absences'].quantile(0.25)
Q3 = df['absences'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

image.png

**The interquartile range is equal to Q3 minus Q1.**

**For example,** 
* consider the following numbers: 1, 3, 4, 5, 5, 6, 7, 11.

    * Q1 is the middle value in the first half of the data set.
    * Since there are an even number of data points in the first half of the data set, the middle value is the average of the two middle values; that is, Q1 = (3 + 4)/2 or Q1 = 3.5. Q3 is the middle value in the second half of the data set. 
    * Again, since the second half of the data set has an even number of observations, the middle value is the average of the two middle values; that is, Q3 = (6 + 7)/2 or Q3 = 6.5.
    * The interquartile range is Q3 minus Q1, so IQR = 6.5 - 3.5 = 3.

## Skewness

In probability theory and statistics, skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. The skewness value can be positive or negative, or undefined
